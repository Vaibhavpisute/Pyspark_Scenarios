```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark=SparkSession.builder\
        .appName("Scenario_6")\
        .master("local[*]")\
        .config("spark.executor.memory","6g")\
        .config("spark.driver.memory","4g")\
        .config("spark.sql.adaptive.enabled","true")\
        .config("spark.sql.shuffle.partitions","50")\
        .config("spark.executor.heartbeatInterval","60s")\
        .config("spark.network.timeout","300s")\
        .config("spark.hadoop.io.native.lib.available","false")\
        .config("spark.memory.fraction","0.25")\
        .config("spark.serializer","org.apache.spark.serializer.KryoSerializer")\
        .getOrCreate()
```

```python
df_events=spark.read\
    .option("inferschema",True)\
    .option("header",True)\
    .option("mergeSchema",True)\
    .parquet("E:/Big_data/Ansh_lamba/Post/Post_3/pyspark_scenarios/dataset_generation/data/dataset2_user_events/events/*/")

df_user=spark.read\
        .option("inferschema",True)\
        .option("header",True)\
        .parquet("E:/Big_data/Ansh_lamba/Post/Post_3/pyspark_scenarios/dataset_generation/data/dataset2_user_events/user_profile/")
```

```python
df_events_1=df_events.join(df_user,"user_id","inner").cache()
df_events_1.count()

df_2=df_events_1.groupBy('country','device_type','user_segment')\
    .agg(round(avg('duration_seconds'),2)\
    .alias('total_avg_duration'))\
    .orderBy('country')

var=df_2.select(col('device_type')).distinct().collect()

for i in range(len(var)):
    a=var[i]['device_type']
    print(a)
    df_4=df_2.filter(col("device_type")==a).orderBy(col('total_avg_duration').desc())
    df_4.limit(5).show()


df_2.groupBy('user_segment').agg(round(avg('total_avg_duration'),2).alias('total_max_time')).show()
```
